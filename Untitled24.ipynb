```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
from matplotlib.colors import ListedColormap
```

```python
X, y = make_classification(n_samples=300, n_features=2, n_redundant=0,
                           n_informative=2, n_clusters_per_class=1, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
svm_linear = SVC(kernel='linear', C=1)
svm_rbf = SVC(kernel='rbf', C=1, gamma=0.5)

svm_linear.fit(X_train, y_train)
svm_rbf.fit(X_train, y_train)
```

```python
print("Linear SVM Accuracy:", svm_linear.score(X_test, y_test))
print("RBF SVM Accuracy:", svm_rbf.score(X_test, y_test))

print("\nConfusion Matrix (Linear SVM):")
print(confusion_matrix(y_test, svm_linear.predict(X_test)))

print("\nClassification Report (Linear SVM):")
print(classification_report(y_test, svm_linear.predict(X_test)))
```

```python
def plot_decision_boundary(clf, X, y, title):
    X_vis = scaler.inverse_transform(X)
    x_min, x_max = X_vis[:, 0].min() - 1, X_vis[:, 0].max() + 1
    y_min, y_max = X_vis[:, 1].min() - 1, X_vis[:, 1].max() + 1
    h = 0.01
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    grid = np.c_[xx.ravel(), yy.ravel()]
    Z = clf.predict(scaler.transform(grid))
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(8, 6))
    plt.contourf(xx, yy, Z, cmap=ListedColormap(('lightcoral', 'lightblue')), alpha=0.7)
    plt.scatter(X_vis[:, 0], X_vis[:, 1], c=y, cmap=ListedColormap(('red', 'blue')), edgecolors='k')
    plt.title(title)
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.grid(True)
    plt.show()
```

```python
plot_decision_boundary(svm_linear, X_train, y_train, "Decision Boundary (Linear SVM)")
plot_decision_boundary(svm_rbf, X_train, y_train, "Decision Boundary (RBF SVM)")

```

```python
cv_scores = cross_val_score(svm_rbf, np.vstack((X_train, X_test)), np.hstack((y_train, y_test)), cv=5)
print("Cross-Validation Accuracy (RBF Kernel):", np.mean(cv_scores))
```
